```python
from collections import deque

# Goal State
goal_state = "123456780"   # 0 is the blank tile

# Possible moves based on index of 0
moves = {
    0: [1, 3],
    1: [0, 2, 4],
    2: [1, 5],
    3: [0, 4, 6],
    4: [1, 3, 5, 7],
    5: [2, 4, 8],
    6: [3, 7],
    7: [4, 6, 8],
    8: [5, 7]
}

def bfs(start_state):
    """Solve 8 Puzzle using BFS"""
    visited = set()
    queue = deque([(start_state, [])])   # (current state, path)

    while queue:
        state, path = queue.popleft()

        # Goal check
        if state == goal_state:
            return path + [state]

        visited.add(state)

        zero_pos = state.index("0")

        for move_pos in moves[zero_pos]:
            new_state = list(state)
            # Swap 0 with the tile
            new_state[zero_pos], new_state[move_pos] = new_state[move_pos], new_state[zero_pos]
            new_state = "".join(new_state)

            if new_state not in visited:
                queue.append((new_state, path + [state]))

    return None


# --- RUN ---
start = "125340678"   # Example initial state

solution = bfs(start)

if solution:
    print("\nSolution Path:")
    for s in solution:
        print(s[0:3], "\n" + s[3:6], "\n" + s[6:9])
        print("-------------")
else:
    print("No solution found!")
```
